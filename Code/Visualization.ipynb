```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a Pandas DataFrame
file_path = 'Clustering_analysis.csv'
data = pd.read_csv(file_path)

# Example 1: Bar Chart
# Create a bar chart to visualize the total number of marginal workers in different age groups.

age_groups = data['Age group']
Industrial_category= data['Cultivators']
plt.figure(figsize=(5, 5))
plt.bar(age_groups, Industrial_category)
plt.title('Cultivators')
plt.xlabel('Age Group')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#import the libraries
import pandas as pd
import matplotlib.pyplot as plt

#providing the dataset to visualise
data = pd.read_csv('industrial_category.csv')
categories = data['Industrial categories']
values = data['total number of workers']

#plot the pie chart
plt.figure(figsize=(8, 18))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=0)
plt.title('Percentage of Workers in Different Industrial Categories')
plt.show()
```
